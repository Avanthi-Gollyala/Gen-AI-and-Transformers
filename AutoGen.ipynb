```python
#GPT Configuration
config_list_gpt = {'config_list':[{'model':'gpt-4o-mini','temperature':0}],}
```

#Building AgenticAI

```python
# Name - Agent "only give exact name of the agent which your creating"/ "meaningful name"
agent_mrk_str = ConversableAgent(name="Marketing_strategist",
                                 system_message="""
                                 you are an expert marketing strategist for both B2B and B2C side of business.
                                 your expertise based in creating strategies for companies and products in the Ed.Tech Industry.
                                 """,
                                 llm_config=config_list_gpt,
                                 human_input_mode='NEVER',
                                 )
```

```python
# Define the task for the agent
content = """ I want some solid marketing strategies for B2B and B2C promotions for my Project Management Course."""
```

```python
reply = agent_mrk_str.generate_reply(messages=[{"content":content, "role":"user"}])
display (Markdown(reply))
```

#Problem Statement: Agentic AI(Two Agent Conversation Agents)
1. You are the Cheif Marketing Officer of an Ed.tech company that sells Project Management class
2. CEO wants to focus on B2C while you(CMO) wants to focus on B2B.
3. Simulate a conversation between CEO and CMO

# Discussion

```python
CMO_agent = ConversableAgent(name="CMO",
                                 system_message="""
                                 you are a Chief Marketing Officer in an Ed.Tech company. You are discussing with the CEO
                                 whether you should focus on B2B or B2C for Project managment courses.
                                 As the CMO, you want to focus on B2B.
                                 """,
                                 llm_config=config_list_gpt,
                                 human_input_mode='NEVER',
                                 )
```

```python
CEO_agent = ConversableAgent(name="CEO",
                                 system_message="""
                                  you are a Chief Executive Officer in an Ed.Tech company. You are discussing with the CMO
                                 whether you should focus on B2B or B2C for Project managment courses.
                                 As the CEO, you want to focus on B2C.
                                 """,
                                 llm_config=config_list_gpt,
                                 human_input_mode='NEVER',
                                 )
```

```python
messages = ["I think we should focus more on B2B business for our project management course."]
```

```python
# Initiate the conversation
result = CMO_agent.initiate_chat(CEO_agent, message=messages[0],
                        max_turns=3, silent=False,
                                 summary_method="reflection_with_llm")
```
